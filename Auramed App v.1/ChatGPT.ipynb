```python
from openai import OpenAI
import httpx
import os
import json

key = os.getenv("OPENAI_API_KEY")

client = OpenAI(api_key=key, http_client=httpx.Client(verify=False))

file = client.files.create(
    file=open("aree_specialistiche_lombardia_Chat.csv", "rb"),
    purpose="user_data"
)

prompt_utente = '''
mi fa male a fare la pipì
'''

prompt_chat = f'''
Agisci come un operatore medico esperto di triage. Sulla base di questa affermazione "{prompt_utente}" definisci se il soggetto è in una situazione critica di pericolo di morte e quale sia il reparto ospedaliero più appropriato.
I nomi dei reparti devono essere presi dal file csv allegato, che contiene una lista di reparti ospedalieri in Lombardia.

Restituisci unicamente un json valido senza markdown e senza testo aggiuntivo, formattato in questo modo: {{"isCritical": "Y", "reparto": "OCULISTICA"}}'''

response = client.responses.create(
    model="gpt-4.1-mini",
    input=[{
        "role": "user",
        "content":
            [{
                "type": "input_text",
                "text": prompt_chat
             },
             
            {
                "type": "input_file",
                "file_id": file.id
            }]
    }]
)

data = json.loads(response.output_text)
print(data['isCritical'])
print(data['reparto'])
```
