```python
import numpy as np
import json
```

```python
def snapshot_ps(n_pazienti_min,n_pazienti_max, score_max):

    codici = ["Rosso", "Arancione", "Azzurro", "Verde", "Bianco"]

    probabilita = [0.10, 0.15, 0.20, 0.40, 0.15]

    n_pazienti = np.random.randint(n_pazienti_min, n_pazienti_max + 1)

    conteggi = np.random.multinomial(n_pazienti, probabilita)
    conteggi = [int(x) for x in conteggi]

    pesi = {
        "Rosso": 20,
        "Arancione": 10,
        "Azzurro": 5,
        "Verde": 2,
        "Bianco": 1
    }

    snapshot = dict(zip(codici, conteggi))

    score = sum(snapshot[c] * pesi[c] for c in snapshot)

    affollamento = min(1.2, (score / score_max))

    return {
        "codici": snapshot,
        "n_pazienti": n_pazienti,
        "affollamento": round(affollamento, 2)
    }



print(snapshot_ps(10, 80, 180))
```

```python
with open("ospedali_docs.json", encoding="utf-8") as f:
    docs = json.load(f)

snapshot_list = []

for doc in docs:

    nome = doc["Nome"]
    classificazione = doc["Classificazione"]

    if "HUB" in classificazione:
        snapshot = snapshot_ps(n_pazienti_min=10, n_pazienti_max=80, score_max=180)
    else:
        snapshot = snapshot_ps(n_pazienti_min=5, n_pazienti_max=40, score_max=100)
    snapshot["nome"] = nome
    snapshot_list.append(snapshot)

for snapshot in snapshot_list:
    print(snapshot)
```

```python
def score_ps(snapshot,tempo,reparto):
    omega_1 = 0.5
    omega_2 = 0.3
    omega_3 = 0.2

    t_max = 90

    T = max(0, 1 - (tempo / t_max))
    A = 1 - snapshot["affollamento"]
    R = 1 if reparto else 0

    score = omega_1 * T + omega_2 * A + omega_3 * R

    return round(score, 2)

for snapshot in snapshot_list:
    print(f"{snapshot['nome']}: {score_ps(snapshot, tempo=0, reparto=False)}")
```
